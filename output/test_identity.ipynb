```python
import pandas as pd
```

```python
df_1500 = pd.read_json("qa_eval_chunk:1500_instruct:False_rerank:True.json")
df_500 = pd.read_json("qa_eval_chunk:700_instruct:False_rerank:True.json")

concat = pd.concat([df_1500, df_500])
```

```python
print(len(df_1500["retrieved_docs"].astype(str).unique()))
print(len(df_500["retrieved_docs"].astype(str).unique()))
print(len(concat["retrieved_docs"].astype(str).unique()))
```

```python
df_1500["len_retrieved"] = df_1500["retrieved_docs"].apply(
    lambda x: sum([len(y) for y in x])
)
df_500["len_retrieved"] = df_500["retrieved_docs"].apply(
    lambda x: sum([len(y) for y in x])
)
```
